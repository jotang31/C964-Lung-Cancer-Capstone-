## Training the Model

### Below, we split the data into independent variables (X) and dependent variables (y). We also do some preprocessing on the column 'GENDER' using Panda's get_dummies() function to change the categorical data ('M' and 'F') into numerical data. This is important because most machine learning models only work with numerical data.

```python
X = df.drop(columns=['LUNG_CANCER']) #indpendent variables
y = df[['LUNG_CANCER']].copy() #dependent variables

X_after_get_dummies = pd.get_dummies(X, columns=['GENDER'])
X_after_get_dummies
#X_after_get_dummies.columns
```

### Now we are ready to split the dataset into 2 parts, one part for training the model and one part for testing the model. Here we choose to reserve 30% of the data for testing. The other 70% will be used for training. Note, the testing data and training data are further split to designate their independent and dependent variables (X and y).

```python
import numpy as np
from sklearn.model_selection import train_test_split

#split the variable sets into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split(X_after_get_dummies, y, test_size=0.3, random_state=12)
```

### Here is a nice display of the split between the independent variables and our dependent variable for lung cancer.

```python
#Nice displays are nice but not required. 
from IPython.display import display_html 
X_train_styler = X_train.head(7).style.set_table_attributes("style='display:inline'").set_caption('Independents variables')
y_train_styler = y_train.head(7).style.set_table_attributes("style='display:inline'").set_caption('Dependents variables')
space = "\xa0" * 10 #space between columns
display_html(X_train_styler._repr_html_()+ space  + y_train_styler._repr_html_(), raw=True)
```

```python
from sklearn import svm, metrics

# SVM Model
svm_model = svm.SVC(gamma='scale', C=1) #Creates a svm model object. Mote, 'scale' and 1.0 are gamma and C's respective defaults 
y_train_array = np.ravel(y_train.values) # Convert the DataFrame into an array to silence warning.
svm_model.fit(X_train,y_train_array) # Train the model

#Using test data
predictions = svm_model.predict(X_test)
score = metrics.accuracy_score(y_test, predictions)
print("first model: ", score)
#90

#Using train data
#predictions = svm_model.predict(X_train)
#score = metrics.accuracy_score(y_train, predictions)
#score
#86 - oh no, it's worse...odd. Normally, we would get better results if we predict using training data because the model has already seen it.

#see the confusion matrix breakdown (makes 93 predictions)
cm = metrics.confusion_matrix(y_test, predictions, labels=svm_model.classes_)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=svm_model.classes_)
disp.plot();
#results: it predicted YES 93 times. Reality: there are 9 NOs and 84 YESs
#a confusion matrix is used for classification models and 
#it describes the performance of a classificaiton madel on test data for which true values are known.


# Let's train using cross-validation
from sklearn.model_selection import KFold, cross_val_score
# Our 'y' is a dataframe but this functions requires the indepenent variables ot be a 1d array.
y_array = np.ravel(y.values)

k_folds = KFold(n_splits = 5, shuffle=True)
# The number of folds determines the test/train split for each iteration. 
# So 5 folds has 5 different mutually exclusive training sets. 
# That's a 1 to 4 (or .20 to .80) testing/training split for each of the 5 iterations.

svm_model_2 = svm.SVC(C=1, random_state = 42) #Creates another svm model object.
scores = cross_val_score(svm_model_2, X_after_get_dummies, y_array)
# This shows the average score. Print 'scores' to see an array of individual iteration scores.
print("Average Score: ", scores.mean())
#87 better than test data

# Logistic Regression Model - let's try and see how the prediction accuracy compares with the SVM model.
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(random_state=0).fit(X_train, y_train_array)
predictions_log = log_model.predict(X_test)
score_log = metrics.accuracy_score(y_test, predictions_log)
print("score_log: ", score_log)
#94.6 Wow

#see the confusion matrix breakdown for logistic regression (makes 93 predictions)
cm_log = metrics.confusion_matrix(y_test, predictions_log, labels=svm_model.classes_)
disp_log = metrics.ConfusionMatrixDisplay(confusion_matrix=cm_log, display_labels=svm_model.classes_)
disp_log.plot();
#results: it predicted YES 85 times, 82 were true positives. 
# Out of 93 predictions, 88 were correct.
```
